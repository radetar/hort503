```python
import numpy as np
import pandas as pd
poplar_annots = pd.read_csv('Ptrichocarpa_444_v3.1.annotation_info.txt', sep='\t')
poplar_annots.head()
```

```python
names = ["Vocabulary", "Term", "Definition"]
go_annots = pd.read_csv('GO.terms.txt', sep = "\t", names= names)
go_annots.head()
```

```python
simple_poplar_annots = poplar_annots.iloc[:,[3,9]]
simple_poplar_annots.head()
```

```python
simple_poplar_annots=simple_poplar_annots.dropna(axis=0, how='any')
simple_poplar_annots.head()
```

```python
go_dfs = []
for index, row in simple_poplar_annots.iterrows():
    terms = row.GO.split(',')
    peptide = np.repeat(row.peptideName, len(terms))
    single=pd.DataFrame(terms,peptide)
    go_dfs.append(single)

```

```python
print(go_dfs[0])
```

```python
whole_poplar = pd.concat(go_dfs)
```
