```python
import pandas as pd
import matplotlib
```

```python
raw_corn = pd.read_csv('corn.csv',)
raw_corn.head(30)
```

```python
corn_bt = raw_corn.iloc[3:17, :19]
corn_bt.head()
```

```python
corn_bt.columns = raw_corn.iloc[1,0:19]
corn_bt.head()
```

```python
melted_corn_bt = corn_bt.melt(id_vars = "State")
melted_corn_bt.head(5)
```

```python
melted_corn_bt.columns = ['State','Year','Value']
melted_corn_bt.head()
```

```python
melted_corn_bt['Crop'] = 'Corn'
melted_corn_bt['Crop Title'] = 'Genetically Engineered (GE) corn'
melted_corn_bt['Variety'] ='Insect-resistant (Bt) only'
melted_corn_bt['Unit'] ='Percentage of all corn planted'
melted_corn_bt = melted_corn_bt[['State','Crop','Crop Title','Variety','Unit','Value']]
melted_corn_bt.head()
```

```python
corn_herb = raw_corn.iloc[3:17, 19:37]
corn_herb.columns = raw_corn.iloc[1,19:37]
corn_herb['State'] =raw_corn.iloc[3:17,:1] 


melted_corn_herb = corn_herb.melt(id_vars = "State")
melted_corn_herb.columns = ['State','Year','Value']
melted_corn_herb['Crop'] = 'Corn'
melted_corn_herb['Crop Title'] = 'Genetically Engineered (GE) corn'
melted_corn_herb['Variety'] ='Herbicide-tolerant only'
melted_corn_herb['Unit'] ='Percentage of all corn planted'
melted_corn_herb = melted_corn_herb[['State','Crop','Crop Title','Variety','Unit','Value']]
melted_corn_herb.head()
```

```python
merged_corn = pd.concat([melted_corn_bt, melted_corn_herb])
merged_corn.head()
```
