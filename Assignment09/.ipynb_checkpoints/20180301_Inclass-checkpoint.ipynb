```python
import pandas as pd
import matplotlib
```

```python
raw_corn = pd.read_csv('corn.csv',)
raw_corn.head()
```

```python
corn_bt = raw_corn.iloc[3:17, :19]
corn_bt.head()
```

```python
corn_bt.columns = raw_corn.iloc[1,0:19]
corn_bt.head()
```

```python
melted_corn_bt = corn_bt.melt(id_vars = "State")
melted_corn_bt.head(5)
```

```python
melted_corn_bt.columns = ['State','Year','Percent']
melted_corn_bt.head(5)
```

```python
melted_corn_bt['Crop Title'] = 'Corn'
melted_corn_bt['Variety'] ='Insect-resistant (Bt) only'
melted_corn_bt['Unit'] ='Percentage planted'
melted_corn_bt.head()
```
